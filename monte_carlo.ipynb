# Estimating $\pi$ with Marbles

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from draw_tools import circle, square
```

```python
INTERVAL = (-2, 4)
ITER = 50000
random.seed(42)

rand_x = np.array([random.uniform(INTERVAL[0], INTERVAL[1]) for _ in range(ITER)])
rand_y = np.array([random.uniform(INTERVAL[0], INTERVAL[1]) for _ in range(ITER)])

dist = np.sqrt(rand_x**2 + rand_y**2)

fig, ax = plt.subplots(figsize=(8, 8))
circle(ax)
square(ax)

col = np.where(dist <= 1, 'c', 
               np.where(
                   np.logical_and(
                       np.logical_and(
                           2 <= rand_x, rand_x <= 3
                       ), 
                       np.logical_and(
                           -0.5 <= rand_y, rand_y <= 0.5
                       )),
                   'r', 'm')
      )

ax.scatter(rand_x, rand_y, c=col)
ax.set_xlim(-2, 4)
ax.set_ylim(-2, 2)
ax.set_axis_off()
ax.set_aspect('equal')

circle_count = sum(col == 'c')
square_count = sum(col == 'r')
ratio = circle_count / square_count
ratio
```
