```python
import random
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
def circle(r=1, c=(0, 0)):
    fig, ax = plt.subplots(figsize=(5,5))

    t = np.arange(0, np.pi * 2.0, 0.01)
    x = r * np.cos(t)
    y = r * np.sin(t)

    ax.plot(x + c[0], y + c[1])
```

```python
circle()
```

```python
def rectangle():
    fig, ax = plt.subplots(figsize=(5,5))
    c = 'b'

    points = np.arange(0, 1, 0.01)
    line = [0 for _ in points]

    ax.plot(line, points, c=c)
    ax.plot(list(map(lambda x: x+1, line)), points, c=c)

    ax.plot(points, line, c=c)
    ax.plot(points, list(map(lambda x: x+1, line)), c=c)
```

```python
rectangle()
```

```python
rand_x = np.array([random.uniform(-2, 5) for _ in range(500)])
rand_y = np.array([random.uniform(-2, 2) for _ in range(500)])

dist = np.sqrt(rand_x**2 + rand_y**2)

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

r = 1
c = (0, 0)
t = np.arange(0, np.pi * 2.0, 0.01)
x = r * np.cos(t)
y = r * np.sin(t)

color = 'k'

rect_x = np.arange(2, 3, 0.01)
rect_y = np.arange(0, 1, 0.01)


#right side
ax1.plot([3 for _ in rect_y], rect_y, c=color)
#left side
ax1.plot([2 for _ in rect_y], rect_y, c=color)

#top
ax1.plot(rect_x, [1 for _ in rect_x], c=color)
#bottom
ax1.plot(rect_x, [0 for _ in rect_x], c=color)

ax1.plot(x + c[0], y + c[1])


col = np.where(dist <= 1, 'c', 
               np.where(
                   np.logical_and(
                       np.logical_and(
                           2 <= rand_x, rand_x <= 3
                       ), 
                       np.logical_and(
                           0 <= rand_y, rand_y <= 1
                       )),
                   'r', 'm')
      )

ax1.scatter(rand_x, rand_y, c=col)
ax1.set_xlim(-2.5, 5.5)
ax1.set_ylim(-2.5, 2.2)

ax1.set_aspect('equal')

circle_count = sum(col == 'c')
square_count = sum(col == 'r')
ratio = circle
ratio_x = np.arange(0, 500, 0.01)

ax2.plot(ratio_x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some mock data
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
